```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the provided Excel file
file_path = 'Staffing Ratio.xlsx'
df = pd.read_excel(file_path, sheet_name='Staffing Ratio - 2021+')

df_clean = df.dropna(subset=['Cohen Clinic'])
df_clean = df_clean[df_clean['Cohen Clinic'] != 'AVERAGES']

df_cumulative = df_clean.groupby('Cohen Clinic').agg({
    'Provided Clinical Hours': 'sum',
    '# Total Staff': 'sum',
    'EOCs Per Clinician': 'sum',
    'Completed EOCs': 'sum'
}).reset_index()


plt.figure(figsize=(12, 8))
sns.barplot(x='Completed EOCs', y='Cohen Clinic', data=df_cumulative, ci=None)
plt.title('Completed EOCs Across Clinics (All Grant Years)')
plt.xlabel('Completed EOCs')
plt.ylabel('Cohen Clinic')
plt.tight_layout()
plt.show()


plt.figure(figsize=(12, 8))
sns.barplot(x='# Total Staff', y='Cohen Clinic', data=df_cumulative, ci=None)
plt.title('Total Staff (All Grant Years)')
plt.xlabel('Total Staff')
plt.ylabel('Cohen Clinic')
plt.tight_layout()
plt.show()
```

```python
# Summing relevant columns for each year for '# Total Staff' and 'Completed EOCs'
df_yearly_sum = df_clean.groupby('Grant Year').agg({
    'Completed EOCs': 'sum',
    '# Total Staff': 'sum'
}).reset_index()

# Line chart for the cumulative sum of Completed EOCs across the years
plt.figure(figsize=(10, 6))
sns.lineplot(x='Grant Year', y='Completed EOCs', data=df_yearly_sum)
plt.title('Completed EOCs Across Grant Years')
plt.xlabel('Grant Year')
plt.xticks(df_yearly_sum['Grant Year']) 
plt.ylabel('Completed EOCs')
plt.tight_layout()
plt.show()

# Line chart for the cumulative sum of Total Staff across the years
plt.figure(figsize=(10, 6))
sns.lineplot(x='Grant Year', y='# Total Staff', data=df_yearly_sum)
plt.title('Total Staff Across Grant Years')
plt.xlabel('Grant Year')
plt.xticks(df_yearly_sum['Grant Year']) 
plt.ylabel('Total Staff')
plt.tight_layout()
plt.show()
```
